# Testing the model

### Importing Libraries

```python
from tensorflow import keras
from keras.models import load_model
from keras.utils import load_img, img_to_array
import numpy as np
model = load_model("gesture.h5") #loading the model for testing
path = "c:\\Users\\Ashwin kumar\\Desktop\\HandGesture Project\\Dataset\\test\\5\\1.jpg"
```

### Plotting the image

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
imgs = mpimg.imread(path)
imgplot = plt.imshow(imgs)
plt.show()
```

```python
#loading of the image
img = load_img(path,
                color_mode='grayscale',
                target_size= (64,64))
x = img_to_array(img)#image to array
x.shape
```

```python
type(x)
```

```python
#changing the shape
x = np.expand_dims(x,axis = 0)
```

```python
x.shape
```

### Predicting our results

```python
pred = np.argmax(model.predict(x),axis=1)#predicting the classes
pred
```

```python
index=['0','1','2','3','4','5']
result=str(index[pred[0]])
result
```

```python
p = []

for i in range(0,6):
    for j in range(0,5):
        path = "c:\\Users\\Ashwin kumar\\Desktop\\HandGesture Project\\Dataset\\test\\"+str(i)+"\\"+str(j)+".jpg"
        img = load_img(path,color_mode = "grayscale",target_size= (64,64))
        x = img_to_array(img)
        x = np.expand_dims(x,axis = 0)
        pred = np.argmax(model.predict(x), axis=-1)
        p.append(pred)

print(p)
```

```python
result = []
index=['0','1','2','3','4','5']
for i in p:
    result.append(index[i[0]])
    
print(result)
```
